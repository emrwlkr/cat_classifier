```python
"""Starter notebook: load one image and inspect its feature vector."""

from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
import sys

# Make the package importable from the repo root
sys.path.insert(0, str(Path.cwd().parent))

from cat_classifier import extract_features

# ── change this path to any image in data/ ────────────────────────────────
img_path = Path("../data/cat_a/photo1.jpg")
# ──────────────────────────────────────────────────────────────────────────

img = Image.open(img_path)
features = extract_features(img_path)
feature_names = ["mean_r", "mean_g", "mean_b", "width", "height", "aspect_ratio"]

print("Image size:", img.size)
print()
for name, val in zip(feature_names, features):
    print(f"  {name:>14}: {val:.3f}")

plt.imshow(img)
plt.axis("off")
plt.title(img_path.name)
plt.show()
```
